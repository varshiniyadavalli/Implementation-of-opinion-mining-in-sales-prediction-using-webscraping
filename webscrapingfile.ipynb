```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
#from requests_html import HTMLSession
```

```python
url ="https://www.flipkart.com/search?q=laptops&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
html = urlopen(url)
```

```python
soup = BeautifulSoup(html,'lxml')
type(soup)
```

```python
title = soup.title
print(title)
```

```python
products=[] #List to store name of the product
prices=[] #List to store price of the product
ratings=[]
displays=[]
discounts=[]
```

```python
for a in soup.findAll(attrs={'class':'_3liAhj'}):
    product=a.find('a', attrs={'class':'_2cLu-l'})
    #for a in soup.findAll('div', attrs={'class':'col col-5-12 _2o7WAb'}):
    t_price=a.find('div', attrs={'class':'_1vC4OE'})
    price=t_price.text.strip()
    #for i in price:
     #   if '₹' in i:
      #      r=i.split('₹',)[0]
       #     prices.append(float(r))
    trim_price=''.join(price.split(','))
    rm_rupee=trim_price.split('₹')
    add_rs_price=rm_rupee[1]
    split_price=add_rs_price.split('E')
    final_price=split_price[0]
    rating=a.find('div', attrs={'class':'hGSR34'})
    #rating=a.find('span',attrs={'class':'_2_KrJI'})
    display=a.find('div',attrs={'class':'_1rcHFq'})
    products.append(product.text)
    prices.append(float(final_price))
    ratings.append(rating.text)
    displays.append(display.text)
print(products)
print(prices)
print(ratings)
print(displays)
```

```python
for b in soup.findAll('a',attrs={'class':'_1Vfi6u'}):
    t_d=b.find('span')
    #discount=t_d.text.strip()
    #for i in price:
     #   if '₹' in i:
      #      r=i.split('₹',)[0]
       #     prices.append(float(r))
    #trim_d=''.join(discount.split(' '))
    #rm_d=trim_d.split('% off')
    #add_d=rm_d[0]
    #split_d=add_d.split(' ')
    #final_d=split_d[0]
    #for tag in t_d:
     #   t_d.find('% off').split()
    discounts.append(t_d.text)
print(discounts)
```

```python
df = pd.DataFrame({'Product Name':products,'Price':prices,'display':displays,'rating':ratings,'discount':discounts}) 
df.to_csv('products.csv', index=False, encoding='utf-8')
```
